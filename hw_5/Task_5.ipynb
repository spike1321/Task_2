```python
#Считайте датасет из файла train.csv
import pandas as pd
import bottleneck as bn
df_pd = pd.read_csv("train.csv")
```

```python
#Посчитайте средний возраст пассажиров и его стандартное отклонение с помощью bottleneck
mean_age = bn.nanmean(df_pd["Age"].values)
std_age = bn.nanstd(df_pd["Age"].values)
print(f"Средний возраст: {mean_age}, Стандартное отклонение: {std_age}")
```

```python
#Считайте датасет из файла Housing.csv
df_housing = pd.read_csv("Housing.csv")
#Для каждого столбца определите оптимальный с точки зрения потребления памяти тип данных
def optimize_dtypes(df):
  for col in df.columns:
    if df[col].dtype == "int64":
      df[col] = pd.to_numeric(df[col], downcast="integer")
    elif df[col].dtype == "float64":
      df[col] = pd.to_numeric(df[col],downcast = "float")
    elif df[col].dtype == "object":
      if len(df[col].unique()) / len(df[col]) < 0.5:
        df[col] = df[col].astype("category")
  return df
  #Поменяйте типы данных столбцов датафрейма на выбранные вами в прошлом пункте и сравните потребление памяти до и после оптимизации
print("До оптимизации:", df_housing.memory_usage(deep=True).sum())
df_optimized = optimize_dtypes(df_housing)
print("После оптимизации:", df_optimized.memory_usage(deep=True).sum())
```
