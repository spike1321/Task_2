```python
#Выведите самое популярное мужское и самое популярное женское имя на корабле

# Фильтрация имен по полу
male_names = df[df['Sex'] == 'male']['Name']
female_names = df[df['Sex'] == 'female']['Name']

# Извлечение имен
male_first_names = male_names.str.extract(r'([^.]*$)')[0]
female_first_names = female_names.str.extract(r'([^.]*$)')[0]

# Поиск самых популярных имен
most_popular_male_name = male_first_names.value_counts().idxmax()
most_popular_female_name = female_first_names.value_counts().idxmax()

# Вывод результата
print(f"Самое популярное мужское имя: {most_popular_male_name}")
print(f"Самое популярное женское имя: {most_popular_female_name}")
```

```python
#Самое популярное мужское и женское имя в каждом классе
print("\nСамые популярные имена по классам:")
for pclass in df['Pclass'].unique():
        print(f"\nКласс: {pclass}")
        class_df = df[df['Pclass'] == pclass]

        # Фильтрация имен по полу для текущего класса
        male_names = class_df[class_df['Sex'] == 'male']['Name']
        female_names = class_df[class_df['Sex'] == 'female']['Name']

        # Извлечение имен
        male_first_names = male_names.str.extract(r'([^.]*$)')[0]
        female_first_names = female_names.str.extract(r'([^.]*$)')[0]

        # Поиск самых популярных имен
        most_popular_male_name = male_first_names.value_counts().idxmax() if not male_first_names.empty else "Нет мужских имен"
        most_popular_female_name = female_first_names.value_counts().idxmax() if not female_first_names.empty else "Нет женских имен"

        # Вывод результата
        print("  Самое популярное мужское имя:", most_popular_male_name)
        print("  Самое популярное женское имя:", most_popular_female_name)
```

```python
#Пассажиры старше 44 лет
print("\nПассажиры старше 44 лет:")
print(df[df['Age'] > 44][['Name', 'Age', 'Sex', 'Pclass']])
```

```python
#Пассажиры младше 44 лет мужского пола
print("\nПассажиры младше 44 лет мужского пола:")
print(df[(df['Age'] < 44) & (df['Sex'] == 'male')][['Name', 'Age', 'Sex', 'Pclass']])
```
