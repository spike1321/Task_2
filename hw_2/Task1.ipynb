```python
class Chocolate:
  def __init__(self, price: float, supplement: str = None):

        if price <= 0:
             raise ValueError('Цена батончика должна быть больше 0.')

        self.price = price
        self.supplement = supplement



  def show_my_chocolate(self):
            if self.supplement:
                print(f'Шоколадный батончик, добавка: {self.supplement}')
            else:
                print(f"Обычный шоколадный батончик")
  def __str__(self):
     return(f'Шоколадный батончик, {self.price} рублей')

try:
         chocolate1 = Chocolate(156.5,"фисташки")
         chocolate2 = Chocolate(125.2)
        # chocolate3 = Chocolate(0.0) #- вызывает исключение

         chocolate1.show_my_chocolate()
         print(chocolate1)
         chocolate2.show_my_chocolate()
         print(chocolate2)

except ValueError as e:
        print(e)
```
